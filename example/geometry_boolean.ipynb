# 基本的な形状操作

Primitive形状を作る（XXX, YYYについては後述）
```
obj0 = cgal.makeXXX(...)
obj1 = cgal.makeYYY(...)
```

Primitive形状を表示する（しなくてもかまわない）
```
di = DrawInterface()
di.addObjects([obj0, obj1])
```

形状を任意に位置に移動させる
```
obj0.translate / obj0.locate / obj0.rotate / obj0.orient / obj0.transform / obj0.newcoords
```

方法1: 新しいオブジェクトを作る（ZZZZについては後述）
```
obj2 = cgal.createByZZZZ(obj0, obj1)
## obj2 が新しく作られる。obj0, obj1は変更されない。
```

方法2: オブジェクトを更新する（ZZZZについては後述）
```
cgal.booleanZZZZ(obj0, obj1)
## obj0 が更新される。obj1は変更されない。
## obj0 = cgal.createByZZZZ(obj0, obj1) とほぼ同じ
```

作ることのできるPrimitive形状(XXX, YYY)

XXX, YYY : Box, Cylinder, Sphere, Cone, Capsule, Torus, Extrusion, Tetrahedron


行うことのできる操作(ZZZZ)

ZZZZ : Union(+), Difference(-), Intersection(*)
